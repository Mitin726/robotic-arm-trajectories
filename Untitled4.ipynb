# vamos a hacer el diseño del algoritmo B.Splines (divide y vencerás)

## para entender esto, necesita estos conceptos base

*   n puntos de control
*   k orden de curva
*   y vector de nodos t: secuencia no decreciente de valores t

## algoritmo de Boor

el algoritmo de Boor calcula un punto en la curva para un parámetro t que es dado, donde usa interpolación recursiva sobre los puntos importantes.

está el caso base y caso recursivo, el base es si r= 0 que r es el nivel de recursión, entonces si r=0 retorna el punto importante directamente. y el caso recursivo los combina los subproblemas más pequeños, lo hace linealmente con un peso de $\alpha$ el cual depende de t y del vector de nodos.

la formula usada es esta

$$P_i^r = (1 - \alpha) \cdot P_{i-1}^{r-1} + \alpha \cdot P_i^{r-1}$$

donde:

$$ \alpha = \frac{(t - t_i)}{(t_{i+k-r} - t_i)} $$

### Estrategia Divide y Vencerás (D&V) del Algoritmo de De Boor

Aquí se ve claro cómo el algoritmo de De Boor usa la idea de "Divide y Vencerás". Es como cuando tienes un problema grande y lo partes en pedacitos más fáciles:

*   **Dividir**: Para encontrar un punto en la curva, el algoritmo lo parte en dos problemas más chiquitos: calcular dos puntos que están un nivel más abajo en la recursión (`Pᵢ₋₁ʳ⁻¹` y `Pᵢʳ⁻¹`).
*   **Vencer**: La parte más sencilla es el caso base. Si ya llegamos al nivel `r=0`, no hay que calcular nada más, simplemente devolvemos el punto de control que toca en ese momento.
*   **Combinar**: Una vez que tenemos los resultados de esos problemas chiquitos, los juntamos haciendo una mezcla. Esa mezcla es una interpolación lineal (como un promedio) entre los dos puntos, pero con un peso `α` que decide cuánto aporta cada uno.

### Explicación del Pseudocódigo `DeBoor(r, i, t, puntos, nodos, k)`

Este es el corazón del algoritmo, la función que realmente calcula un punto en la curva B-Spline. Imagina que le das todos los ingredientes y ella te devuelve el punto exacto:

*   **Entradas**: Le pasamos un montón de cosas: `r` (el nivel actual en el que estamos trabajando, que va bajando), `i` (qué punto de control estamos mirando), `t` (el valor de 0 a 1 que nos dice dónde queremos el punto en la curva), la lista de `puntos` de control, la lista de `nodos` y `k` (el orden de la curva).
*   **Caso Base (`SI r == 0`)**: Si `r` ya es `0`, significa que ya llegamos al final de la recursión, y lo único que hacemos es devolver el punto de control `puntos[i]`. ¡Listo!
*   **Cálculo de `α`**: Si no es el caso base, necesitamos calcular `α`. Esto se hace con una fórmula que usa `t` y algunos valores de los `nodos`. Si el `denominador` de la fórmula da `0` (que puede pasar si hay nodos repetidos), para que no haya errores, se pone `α` en `0`.
*   **Divide y Vencerás (Recursión)**: Aquí es donde llamamos a la misma función `DeBoor` dos veces, pero con `r-1` (un nivel menos) y con `i-1` o `i`. Esto crea los `subproblemas` de los que hablamos antes (`izquierda` y `derecha`).
*   **Combinar y Salir (`RETORNAR`)**: Cuando esas dos llamadas recursivas nos devuelven sus resultados, usamos el `α` que calculamos para mezclarlos y devolver el punto final. Así, cada llamada va subiendo los resultados hasta que la primera llamada (`DeBoor` principal) nos da el punto que queríamos.

```
Algoritmo: DeBoor(r, i, t, puntos, nodos, k)

Entrada:
  r     → nivel de recursión actual (empieza en k-1)
  i     → índice del punto de control activo
  t     → parámetro en [0, 1]
  puntos → arreglo de puntos de control
  nodos  → vector de nodos
  k     → orden de la B-Spline

Salida:
  Punto en la curva B-Spline para el parámetro t

INICIO
  // Caso base
  SI r == 0 ENTONCES
      RETORNAR puntos[i]

  // Calcular peso de interpolación
  denominador ← nodos[i + k - r] - nodos[i]
  
  SI denominador == 0 ENTONCES
      α ← 0
  SINO
      α ← (t - nodos[i]) / denominador

  // Divide y Vencerás: resolver dos subproblemas
  izquierda ← DeBoor(r-1, i-1, t, puntos, nodos, k)
  derecha   ← DeBoor(r-1, i,   t, puntos, nodos, k)

  // Combinar resultados
  RETORNAR (1 - α) * izquierda + α * derecha
FIN
```

### Explicación del Pseudocódigo `GenerarTrayectoriaBSpline(puntos, k, pasos)`

Esta función es la que te permite dibujar toda la curva, no solo un punto. Es como si fuera la orquestadora que usa el algoritmo de De Boor muchas veces para ir creando la línea completa:

*   **Inicio**: Primero, averigua cuántos `puntos` de control tienes (`n`).
*   **`GenerarNodosUniformes`**: Luego, crea el `vector de nodos` (la lista `nodos`). Aquí dice `clamped`, que es un tipo de nodos que hace que la curva empiece y termine en los puntos de control de los extremos. Esto es importante para que la curva tenga sentido visualmente.
*   **`trayectoria`**: Crea una lista vacía que se llama `trayectoria`, donde va a guardar todos los puntos que va calculando para formar la curva.
*   **Bucle para `t`**: Aquí viene la magia para dibujar la curva. Va a ir probando muchos valores de `t` (desde `0` hasta `1`), con pequeños `pasos` para que la curva quede suave.
*   **`EncontrarSegmento`**: Para cada `t`, primero averigua en qué parte de la curva estamos (`i`). Esto es importante para saber qué `nodos` y `puntos` de control son relevantes.
*   **Llamar a `DeBoor`**: Con ese `t` y `i` (y los otros datos), llama a la función `DeBoor` que explicamos antes para que le calcule un `punto` de la curva.
*   **Guardar Punto**: Ese `punto` calculado lo añade a la lista `trayectoria`.
*   **Fin**: Cuando termina de probar todos los `t`, devuelve la `trayectoria` completa, que es la curva B-Spline.

```
Algoritmo: GenerarTrayectoriaBSpline(puntos, k, pasos)

INICIO
  n ← longitud(puntos)
  nodos ← GenerarNodosUniformes(n, k)   // knot vector clamped
  trayectoria ← []

  PARA cada t en [0, 1] con 'pasos' divisiones:
      i ← EncontrarSegmento(t, nodos, k)    // índice del nodo activo
      punto ← DeBoor(k-1, i, t, puntos, nodos, k)
      Agregar punto a trayectoria

  RETORNAR trayectoria
FIN
```

## Análisis de Complejidad del Algoritmo de De Boor

Vamos a analizar la complejidad temporal del algoritmo `DeBoor` y de la función `GenerarTrayectoriaBSpline`.

### 1. Relación de Recurrencia para `DeBoor`

En cada llamada a `DeBoor(r, i, t, puntos, nodos, k)`, si `r > 0`, se hacen dos llamadas recursivas a `DeBoor(r-1, ...)` y el trabajo realizado en cada nivel es constante: calcular `α` y la interpolación lineal. El "tamaño" del problema se puede considerar `r` (el nivel de recursión).

Esto nos da la siguiente relación de recurrencia:

$$T(r) = 2T(r-1) + O(1)$$

Donde `r` es el nivel de recursión actual, que va desde `k-1` (inicio) hasta `0` (caso base).

### 2. Árbol de Recursividad

Podemos visualizar la ejecución del algoritmo como un árbol de recursividad:

*   **Nivel 0:** `DeBoor(r)` → 1 nodo, trabajo O(1)
*   **Nivel 1:** `DeBoor(r-1)` `DeBoor(r-1)` → 2 nodos, trabajo O(1) c/u
*   **Nivel 2:** `DeBoor(r-2)` ... `DeBoor(r-2)` → 4 nodos, trabajo O(1) c/u
*   ...
*   **Nivel `r`:** (caso base, `r=0`) → `2^r` nodos, trabajo O(1) c/u

La altura del árbol es `r`. El número total de nodos (llamadas a la función) es la suma de nodos en cada nivel:

$$T(r) = \sum_{i=0}^{r} 2^i = 2^{r+1} - 1 = O(2^r)$$

Dado que el nivel inicial de recursión es `k-1` (siendo `k` el orden de la curva), la complejidad del algoritmo `DeBoor` para calcular un punto es:

$$\boxed{T(k) = O(2^k)}$$

### 3. Método Maestro (para Restas)

La relación de recurrencia es $T(r) = 2T(r-1) + O(1)$.

Esta forma no se ajusta directamente al Teorema Maestro clásico ($T(n) = aT(n/b) + f(n)$) porque la recursión es sobre `n-1` en lugar de `n/b`.

Sin embargo, podemos usar una variante del Teorema Maestro para recurrencias de la forma:

$$T(n) = aT(n-b) + f(n)$$

Para nuestra recurrencia: `a = 2`, `b = 1`, y `f(n) = O(1) = O(n^0)`.

Si $a > 1$, la solución es $T(n) = O(a^{n/b} \cdot f(n))$. Aplicando esto:

$$T(r) = O(2^{r/1} \cdot O(1)) = O(2^r)$$

Confirmando el resultado obtenido por el árbol de recursividad, y sustituyendo `r` por `k-1` (que es el orden efectivo de la recursión):

$$\boxed{T(k) = O(2^k)}$$

Esto significa que la complejidad del algoritmo De Boor es **exponencial** con respecto al orden `k` de la curva. Afortunadamente, `k` suele ser un número pequeño y fijo (e.g., 3 para curvas cuadráticas, 4 para cúbicas), por lo que no crece con el número de puntos de control `n`.

### 4. Complejidad Total de la Generación de la Trayectoria (`GenerarTrayectoriaBSpline`)

La función `GenerarTrayectoriaBSpline` calcula `pasos` puntos en la curva. Para cada punto, llama al algoritmo `DeBoor`. Las operaciones de `EncontrarSegmento` y `GenerarNodosUniformes` generalmente tienen una complejidad que depende linealmente del número de puntos de control `n` o del número de nodos (`n+k+1`), pero se realizan una vez o `pasos` veces con una complejidad baja.

La complejidad dominante viene de las `pasos` llamadas a `DeBoor`:

$$T_{total} = \text{pasos} \times T(k) = O(\text{pasos} \times 2^k)$$

Considerando que `k` es un valor constante (ya que el orden de la curva es típicamente fijo y no crece con la entrada principal `n`):

*   Si `pasos` es proporcional al número de puntos de control `n` (por ejemplo, para mantener una densidad de puntos similar a medida que `n` crece), entonces:

    $$\boxed{T_{total}(n) = O(n \cdot 2^k)} \quad \text{con k constante}$$

*   Si `pasos` es un valor fijo e independiente de `n` (por ejemplo, siempre se generan 100 puntos), entonces la complejidad sería simplemente $O(2^k)$, lo cual es constante en términos de `n`.

En la mayoría de las implementaciones prácticas donde `k` es pequeño, la complejidad lineal en `n` (el número de puntos de control) y en `pasos` es aceptable.

Importamos las librerias

````python
import numpy as np
import matplotlib.pyplot as plt
import time
````

### 1. Función para Crear los Nodos (Knot Vector)

````python
# 1. Crear los nodos (knot vector)
def crear_nodos(n, k):
    nodos = []
    total = n + k

    for i in range(total):
        if i < k:
            nodos.append(0.0)
        elif i >= n:
            nodos.append(1.0)
        else:
            valor = (i - k + 1) / (n - k + 1)
            nodos.append(valor)

    return nodos
````

### 2. Función para Buscar el Tramo de `t`

````python
# 2. Buscar en qué parte cae t
def buscar_tramo(t, nodos, k, n):
    if t == 1.0:
        i = n - 1
        while i > 0 and nodos[i] == 1.0:
            i -= 1
        return i

    for i in range(k - 1, n):
        if nodos[i] <= t < nodos[i + 1]:
            return i

    return n - 1
````

### 3. Implementación del Algoritmo de De Boor (Recursivo)

````python
# 3. Algoritmo de De Boor (recursivo)
def deboor(r, i, t, pts, nodos, k):
    # caso base
    if r == 0:
        return np.array(pts[i])

    den = nodos[i + k - r] - nodos[i]

    if den == 0:
        a = 0
    else:
        a = (t - nodos[i]) / den

    izq = deboor(r - 1, i - 1, t, pts, nodos, k)
    der = deboor(r - 1, i, t, pts, nodos, k)

    return (1 - a) * izq + a * der
````

### 4. Función para Generar la Curva B-Spline Completa

````python
# 4. Generar toda la curva
def hacer_bspline(puntos, k, pasos=100):
    n = len(puntos)
    nodos = crear_nodos(n, k)
    curva = []

    for p in range(pasos + 1):
        t = p / pasos
        i = buscar_tramo(t, nodos, k, n)
        punto = deboor(k - 1, i, t, puntos, nodos, k)
        curva.append(punto)

    return curva
````

### 5. Demostración y Dibujo de la Curva

````python
# 5. Probar y dibujar
puntos = [(0,0), (2,5), (5,8), (8,5), (10,0), (8,-5), (5,-8)]
k = 3

curva = hacer_bspline(puntos, k, pasos=200)

x_p = [p[0] for p in puntos]
y_p = [p[1] for p in puntos]
x_c = [p[0] for p in curva]
y_c = [p[1] for p in curva]

plt.figure()
plt.plot(x_c, y_c, label="B-Spline")
plt.scatter(x_p, y_p)
plt.plot(x_p, y_p, '--')
plt.title("Curva B-Spline")
plt.legend()
plt.grid()

plt.savefig("bspline.png")
plt.close()

print("Imagen de la curva guardada")
````

### 6. Medición de Tiempos de Ejecución

````python
# 6. Medir tiempos
def tiempo_prueba(n, k, pasos=100):
    pts = [(i, np.sin(i)) for i in range(n)]

    inicio = time.perf_counter()
    hacer_bspline(pts, k, pasos)
    fin = time.perf_counter()

    return fin - inicio
````

### 7. Pruebas de Rendimiento y Gráfico de Tiempos

````python
valores = list(range(5, 30, 2))
k_fijo = 3

tiempos = []
for n in valores:
    t = tiempo_prueba(n, k_fijo)
    tiempos.append(t)

# línea tipo O(n) solo para comparar
base = tiempos[0] / valores[0]
aprox = [base * n for n in valores]

plt.figure()
plt.plot(valores, tiempos, 'o-', label="Tiempo real")
plt.plot(valores, aprox, '--', label="Aprox O(n)")
plt.xlabel("n")
plt.ylabel("tiempo")
plt.title("Tiempo vs n")
plt.legend()
plt.grid()

plt.savefig("tiempos.png")
plt.close()

print("Gráfica de tiempos guardada")
````
